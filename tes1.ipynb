```python
import tensorflow as tf
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, LabelEncoder, StandardScaler
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
```

```python
def import_data():
    df = pd.read_csv("dataset/1/train.csv")
  
    x=df[["age", "workclass", "fnlwgt", "educational-num", "gender","hours-per-week"]]
    y=df['income_>50K']
    return x, y

input_dataset, output_dataset = import_data()
```

```python
input_dataset['gender'] = LabelEncoder().fit_transform(input_dataset['gender'])
input_dataset["workclass"] = LabelEncoder().fit_transform(input_dataset["workclass"].astype(str))
SS = StandardScaler()
input_dataset = SS.fit_transform(input_dataset)
x_train, x_test, y_train, y_test = train_test_split( input_dataset, output_dataset, test_size = 0.3, random_state = 0) 
```

```python
y_train
```

```python
models = tf.keras.models.Sequential()
models.add(tf.keras.layers.Dense(6, activation='relu', input_shape = (6, )))
models.add(tf.keras.layers.Dense(2, activation='relu'))
models.add(tf.keras.layers.Dense(1, activation='sigmoid'))
models.compile(optimizer=tf.keras.optimizers.Adam(0.001), loss="binary_crossentropy", metrics = ['accuracy'])
```

```python
models.fit(x_train, y_train, epochs=300)
```
